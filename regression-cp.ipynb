```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
# from sklearn.model_selection import KFold
# from sklearn.model_selection import RepeatedKFold
# import sklearn.metrics
# import datetime
import pandas as pd
# import matplotlib.pyplot as plt
# import numpy as np
from IPython.display import display
```

```python
df = pd.read_csv('data/ys1a.csv') # load yield strength data
display(df)
df = df[df.ys.notnull()] # filter out rows with ys = null
df
```

```python
myset = df[["vec", "delta", "deltachi", "deltahmix", "deltasmix", "ys"]] # filter out non-using features
myseed = 12356789 # for randomizing

# split data into train, test, calibration sets
traincalset, testset, traincalout, testout = train_test_split(myset, myset.ys, train_size=0.85, test_size=0.15, random_state=myseed)
trainset, calset, trainout, calout = train_test_split(traincalset, traincalout, train_size=0.825, test_size=0.175, random_state=myseed)

display(trainset.shape, calset.shape, testset.shape)
display(traincalout - traincalset.ys)
```

```python
LR = LinearRegression() # Instantiate a Linear Regression model
model = LR.fit(trainset.drop('ys', axis=1), trainout) # Train the Linear Regression model. trainset is the training feature set and trainout is the corresponding target variable
```

```python
y_cal_pred = model.predict(calset.drop('ys', axis=1)) # predict on calibration set
display(y_cal_pred)

y_cal_error = abs(y_cal_pred - calset.ys) # calculate error
y_cal_error
```

```python
quantile = y_cal_error.quantile(q=.95, interpolation='midpoint') # calculate 1-alpha quantile from the distribution of absolute errors
quantile
```

```python
y_test_pred = model.predict(testset.drop('ys', axis=1)) # predict on test set
y_test_pred
```

```python
# calculate prediction interval
y_test_interval_pred_left = y_test_pred - quantile
y_test_interval_pred_right = y_test_pred + quantile

index = testout.index  # get index of a df
coverage = [
    1 if y_test_interval_pred_left[i] <= testset.ys[index[i]] <= y_test_interval_pred_right[i] else 0
    for i in range(len(index))
]  # Determine coverage for each prediction in the test set. If the actual value falls within the prediction interval, set coverage to 1, otherwise set it to 0

# Create a df containing information about the actual values, prediction intervals, predicted values, and coverage
interval = pd.DataFrame(
    list(zip(testout, y_test_interval_pred_left, y_test_interval_pred_right, y_test_pred, coverage)),
    columns=["actual", "lower_interval", "upper_interval", "predicted", "coverage"],
)
interval
```

```python
print("Percentage of coverage", sum(interval.coverage)/len(interval.coverage))
```
